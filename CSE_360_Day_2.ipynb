# ANN Model

Employee Retention dataset

```python
#import the libraries

import numpy as np
import pandas as pd

df = pd.read_csv("/content/HR_comma_sep - HR_comma_sep.csv")
```

```python
df
```

```python
df1 =["sales","salary"]
df_final = pd.get_dummies(df, columns=df1, drop_first=True)
```

```python
df_final
```

```python
#spliting x and y variable

X = df_final.drop(['left'], axis =1).values
y = df_final['left'].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(9, kernel_initializer="uniform", activation = 'relu', input_dim = 18))
model.add(Dense(5, kernel_initializer="uniform", activation ='relu'))
model.add(Dense(1, kernel_initializer= "uniform", activation ="sigmoid"))
model.compile(optimizer='adam', loss="binary_crossentropy", metrics =["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=150)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = y_pred.round()
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
ann = Sequential()

# Input Layer
ann.add(Dense(12, input_dim=11,kernel_initializer='uniform',activation='relu'))

# Hidden layer
ann.add(Dense(8,activation='relu'))

# Output layer
ann.add(Dense(1,activation='sigmoid'))
```

```python
# compile the model
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

# Fit the model
res = ann.fit(X_train,y_train,batch_size=32,epochs= 200)
```

```python
y_pred = np.round(ann.predict(X_test)).astype(int)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

# CNN model

```python
from keras.datasets import mnist
from keras.layers import Dropout, Conv2D, MaxPool2D, Flatten
from keras.utils import np_utils
```

```python
#to calculate accuracy
from sklearn.metrics import accuracy_score

# loading the dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# building the input vector from the 28x28 pixels
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28,1)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
#normalizing the data to help with the training
X_train /= 255
X_test /= 255
```

```python
# One hot encoding using keras numpy related utilities
n_classes = 10
print("before encoding:",y_train.shape)
y_train = np_utils.to_categorical(y_train,n_classes)
y_test = np_utils.to_categorical(y_test,n_classes)
print("After encoding:",y_train.shape)
```

```python
# building a linear stack of layers with the sequential model
model = Sequential()
#convultional layer
model.add(Conv2D(32, kernel_size=(3,3), strides=(1,1), padding='valid', activation='relu', input_shape =(28, 28, 1)))
model.add(MaxPool2D(pool_size=(1,1)))
#flatten output of conv
model.add(Flatten())
#hidden layer
model.add(Dense(100, activation='relu'))
#output layer
model.add(Dense(10, activation ='softmax'))

#compiling the sequential model
model.compile(loss = 'categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
#training the model for 10 epochs
model.fit(X_train, y_train, batch_size=120, epochs=10, validation_data=(X_test, y_test))
```
