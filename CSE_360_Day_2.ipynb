```python

#import the libraries

import numpy as np
import pandas as pd

df = pd.read_csv("/content/HR_comma_sep - HR_comma_sep.csv")
```

```python
df
```

```python
df1 =["sales","salary"]
df_final = pd.get_dummies(df, columns=df1, drop_first=True)
```

```python
df_final
```

```python
#spliting x and y variable

X = df_final.drop(['left'], axis =1).values
y = df_final['left'].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(9, kernel_initializer="uniform", activation = 'relu', input_dim = 18))
model.add(Dense(5, kernel_initializer="uniform", activation ='relu'))
model.add(Dense(1, kernel_initializer= "uniform", activation ="sigmoid"))
model.compile(optimizer='adam', loss="binary_crossentropy", metrics =["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=150)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = y_pred.round()
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
